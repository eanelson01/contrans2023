# Lingo and Rules
* Entity - Table
* Attribute - Column
* Primary Key - One or more columns in a table that are unique on every row

Build in first normal form, then second, then third. Go in order and don't skip steps

###  First normal form:
* Every table must have a primary key
* Values in every cell in every table must be atomic (no lists or dicitonaries)
* There are no repeating groups
    * A problem when there is non-atomic data
 
### Second Normal Form:
* Data is in 1NF
* Attributes must depend on the whole primary key, not just a subset of the key
* Depends means one values of the primary key maps to one value of the attribute

### Third Normal Form:
* Be in 2NF
* Every attribute is not functionally dependent on everything outside of the primary key

### Pnemonic to remember 1NF, 2NF, 3NF:
The key, the whole key, nothing but the key, so help me Codd
